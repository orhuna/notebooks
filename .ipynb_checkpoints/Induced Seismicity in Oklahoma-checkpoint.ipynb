# Understanding Water-Injection Patterns in Oklahoma
Waste-water disposal is linked to induced seismicity in Oklahoma by various authors (Keranen et al., 2013; Walsh and Zoback, 2015; Walsh and Zoback, 2016). In this demo, we will explore characteristics of the injection patterns of water-disposal wells in OK.

One method of understanding characteristics of time signal such as a well-injection rate is through decomposition. In this example, we will use a R library to analyze temporal patterns in our data and incorporate it in our spatial analysis. In particular, we will use functional data analysis library of R. Functional data analysis states that any given complex signal can be decomposed into a series of simpler, analytically-defined signals. 

Functional data analysis does not exists in Pro, yet. However, using the [R-ArcGIS Bridge](https://r-arcgis.github.io/)  you can integrate R libraries into a geoprocessing workflow.

The R-ArcGIS bridge (will be referred to as **the bridge** from now on) is a R library developed by the R-Bridge team (that's us!). The library is called ```arcgisbinding```.

````r
#### IMPORT R-ARCGIS BINDING ####
library(arcgisbinding)
#### IMPORT OTHER R LIBRARIES ####
library(sp, quietly = TRUE)
library(fda, quietly = TRUE)
library(reticulate)
````

After importing the ```arcgisbinding``` library, ```arc.check_product()``` is needed to establish a connection with a valid version of licensed ArcGIS Pro or   Desktop.

````r
#### RUN R-ARCGIS BINDING FUNCTION ####
arc.check_product()
````

```arc``` family are functions that reside in ```arcgisbinding``` library. One of the key functions is ```arc.open``` that establishes a connection between **features classes** or **rasters** and R. Once a feature class is given as input to ```arc.open``` it creates an object that allows interacting with that feauture or raster.  

````r
#### READ FEATURES FROM A FILE GEODATABASE ####
open.obj <- arc.open('C:/Users/orhu8849/Desktop/Chevron/OK_seismicity.gdb/OK_AIC_II')
````

Next step is reading the data into R using ```arc.select```. This function can also perform subsetting via SQL-based commands.

````r
## Bring In the Whole Feature Class
sel.obj <- arc.select(open.obj)

## Visualize Well Locations
plot(sel.obj$Longitude, sel.obj$Latitude, xlab = "West Longitude", ylab = "North Latitude")
````

As you can see aboe all of the well locations are read into R using the R-ArcGIS Bridge. Now the data is in R and we can perform statistical analysis on this data. 

First, we will aggregate this space-time data into a data-structure that is easy to handle within R. For this reason we use aggregate to aggregate water volume injected per day by well locations that are indexes by ```API_NUM```.

````r
#### Aggregate Space Time Date per Well ####
data.well <- aggregate(sel.obj$Volume_BPD, by = list(sel.obj$API_NUM, sel.obj$Report_time), FUN = mean)

colnames(data.well) <- c('WELL.ID', 'TIME', 'VOL_INJ')
head(data.well)
````

````r
well.id.list <- unique(data.well$WELL.ID)

plot(data.well[data.well$WELL.ID == well.id.list[1], 2], data.well[data.well$WELL.ID == well.id.list[1], 3], type = "l", col = 'black', yaxt='n', xaxt='n', ann=FALSE)
par(new = TRUE)
plot(data.well[data.well$WELL.ID == well.id.list[2], 2], data.well[data.well$WELL.ID == well.id.list[2], 3], type = "l", col = 'blue', yaxt='n', xaxt='n', ann=FALSE)
par(new = TRUE)
plot(data.well[data.well$WELL.ID == well.id.list[3], 2], data.well[data.well$WELL.ID == well.id.list[3], 3], type = "l", col = 'red', xlab = "Time", ylab = "V BPD Injected")
````

````r
basis <- create.bspline.basis(norder = 8)
plot(basis)
````

````r
ts <- data.well[data.well$WELL.ID == well.id.list[1], 3]
fda.obj <- Data2fd(y = ts, basisobj = basis)

plot(data.well[data.well$WELL.ID == well.id.list[1], 2], data.well[data.well$WELL.ID == well.id.list[1], 3], type = "l", xlab = "Time", col = 'black', yaxt='n', ann=FALSE)
par(new = TRUE)
plot(fda.obj, main = "fit", col = 'red', lwd = 2, xaxt='n', ylab = "V BPD Injected")
````

````r
print(fda.obj$coef)
````

````r
fda.coefs <- matrix(0L, ncol = 20, nrow = length(well.id.list))

sp.df <- arc.data2sp(sel.obj)

coords <- matrix(nrow = length(well.id.list), ncol = 2)

count = 0

for (ind in well.id.list) {
    count = count + 1
    ts <- data.well[data.well$WELL.ID == ind, 3]

    loc.ind <- which(sp.df$API_NUM == ind)[1]
    coords[count, 1:2] <- sp.df@coords[loc.ind, 1:2]

    fda.obj <- Data2fd(y = ts, basisobj = basis)
    fda.coefs[count, 1:length(fda.obj$coefs)] <- fda.obj$coefs

}
````

````r
points.write <- SpatialPointsDataFrame(coords, as.data.frame(fda.coefs), coords.nrs = sp.df@coords.nrs, proj4string = sp.df@proj4string,bbox = sp.df@bbox)

arc.write('C:/Users/orhu8849/Desktop/Chevron/OK_seismicity.gdb/OK_AIC_decomposed', points.write, overwrite = TRUE)
````

You can call any geoprocessing function from R. In this example we have missing data and we will fill the missing values using spatio-temporal neighbor's average using an ArcGIS tool named Fill Missing Values.

````r
root <- 'C:/Users/orhu8849/Desktop/UC 2018/Pre-UC Seminar/R-Bridge/Climate/Climate.gdb'

weatherStation <- 'StationData'

station.open.obj <- arc.open(file.path(root, weatherStation))

station.data.original <- arc.select(station.open.obj, c( "T_MONTHLY_MEAN", "TIME_RECORDED"))

print(station.data.original)
````

````r
ARCPY$Delete_management(file.path(root, 'StationDataFilled'))

ARCPY$stpm$FillMissingValues(file.path(root, weatherStation), file.path(root, 'StationDataFilled'), "T_MONTHLY_MEAN;T_MONTHLY_AVG", "AVERAGE", "K_NEAREST_NEIGHBORS", NULL, "1 Months", "TIME_RECORDED", 20, "WBANNO", NULL, NULL, NULL, NULL, -9999, NULL)

station.data.filled <- arc.select(arc.open(file.path(root, 'StationDataFilled')), c("T_MONTHLY_MEAN", "TIME_RECORDED"))

print(station.data.filled)
````
